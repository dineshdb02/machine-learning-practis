```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Quote-Equity-ITC-EQ-18-04-2023-to-18-04-2024.csv')
df
```

```python
df.drop(["series "],axis=1,inplace=True)
```

```python
df.set_index('Date ',inplace=True)
```

```python
df
```

```python
for j in df.columns:
    df[j] = df[j].apply(lambda i: float(i.replace(",", "")) if isinstance(i, str) else i)
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.heatmap(df)
```

```python
sns.clustermap(df)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

```python
wcss = []
for i in range(1, 250):
  model= KMeans(n_clusters=i)
  model.fit(df)
  wcss.append(model.inertia_)
```

```python
sns.set_style('whitegrid')
plt.plot(range(1, 25), wcss[:24])
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=5)
model.fit(df)
identify=model.fit_predict(df)
df['cluster']=identify
df
```
